# Lab 1 Preparation: Explanation 

## What is this lab about?

This lab uses a regression model to predict house prices from tabular data.
We give the machine a table of houses and their features, and we ask:
“Can you guess the price without embarrassing yourself?”

## What is a CSV?

A CSV is a plain-text table where each row is a sample and each column is a variable. Pandas can load it with read_csv. Pandas documents read_csv as the standard way to read comma-separated values into a DataFrame.
So it is basically an Excel sheet that went on a diet and lost all the fancy formatting.

## What is the dataset here?

The notebook reads BostonHousing.csv, turns the rows into tensors, uses the last column as the target, and the other columns as features. It reports 13 features.
Basically: 
- Each row is one house.
- The first 13 columns are clues.
- The last column is the answer we want to predict.

## What is a Dataset?

A Dataset is the object that stores samples so the model can access them during training. We can say that the Dataset is the fridge that stores all the ingredients. 

## What is a DataModule?

Lightning says a LightningDataModule is a convenient way to organize training, validation, testing, prediction loaders, and data preparation.
Following the previous analogy: 
The DataModule is the waiter.
It knows which food goes to practice, which goes to checking, and which goes to the final exam.

## What is a LightningModule?

Lightning describes a LightningModule as the place where you organize the model, the training step, validation step, test step, prediction step, and optimizer.

## What is a Trainer?

The Trainer runs the training loop for the model using the data you prepared. In the original notebook it is used together with early stopping and validation tracking.

The Trainer is basically a sadistic gym coach.
It says:
“Again.”
“Again.”
“Again.”
“Stop. You’re not improving anymore.”

## Why split into train, validation, and test?

The README says we (students) should understand why train, validation, and test sets exist, and the notebook uses all three. The notebook splits the dataset into 60% train, 30% validation, and 10% test.

- Train = study with the textbook
- Validation = mini mock exam during studying
- Test = final exam you do not cheat on

## What is an epoch?

An epoch is one full pass over the training data. The notebook explains training as repeatedly going through all train data and then checking validation loss.
One epoch = one full lap around the track.

## What is R²?

Scikit-learn defines R² as the coefficient of determination. Best possible score is 1.0; a constant model that always predicts the mean gets 0.0; and scores can even be negative if the model is worse than that baseline.

R² asks:
“How much of the price pattern did the model actually understand?”

## What is RMSE?

RMSE is root mean squared error, a regression error metric that measures how far predictions are from the true values, in the same unit as the target. Scikit-learn documents it as a standard regression loss.

So it basically asks:
“On average, how badly are we missing the house price?”
